```python

def rain_condition(v):
    """
        Depends of the watter volume there are 4 possible values
        < 1.00:exra dry
        < 1.75: dry
        2.75:"Rain"
        else "heavy rain"
        
    """
    if v < 1.00:
        return "exra dry"
    elif v < 1.75:
        return "dry"
    elif v < 2.75:
        return "Rain"
    return "heavy rain"
        
```

```python
def df_style(styler):
    """
        
        set the style of cell for the resutls
        
    """
    styler.format(rain_condition)
    styler.format_index(lambda v: v.strftime("%A"))
    styler.background_gradient(axis=None, vmin=0.5, vmax=5, cmap = "YlGnBu")
    return styler                      
    
    
```
